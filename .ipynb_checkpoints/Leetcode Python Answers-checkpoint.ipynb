```python
# Top Interview Questions - Easy
# Remove Duplicates from Sorted Array
'''
def removeDuplicates(self, nums: List[int]) -> int:
    i = 0
    j = 0
    while j < len(nums)-1:
        j += 1
        if nums[i]!=nums[j]:
            i+=1
            nums[i]=nums[j]

    return i+1        
'''
# Key points: two-pointer, j to go through the whole list,
# use i to replace the first-k elements

# Best Time to Buy and Sell Stock II
'''
def maxProfit(self, prices: List[int]) -> int:
    maxprofit = 0
    i = 0
    while i < len(prices)-1:
        if prices[i+1]>=prices[i]:
            maxprofit += (prices[i+1]-prices[i])
        i += 1
    return maxprofit
'''
# Key points: figure out that we need to add all the diff between consecutive numbers

# Rotate Array
'''
def rotate(self, nums: List[int], k: int) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    l = len(nums)
    k = k%l
    if k != 0:
        add_list = nums[-k:]
        nums[k:] = nums[:len(nums)-k]
        nums[:k] = add_list
'''
# Key points: only do the operation when k!=0,
# k can be larger than len(nums), then only get the remainder of the division as the new k


# Contains Duplicate
'''
def containsDuplicate(self, nums: List[int]) -> bool:
    nums_set = set(nums)
    return len(nums)!=len(nums_set)
'''
# Key points: use set()

#  Single Number
'''
def singleNumber(self, nums: List[int]) -> int:
    nums_dict = {}
    for num in nums:
        if num not in nums_dict:
            nums_dict[num]=1
        else:
            nums_dict[num]+=1
    for value in nums_dict:
        if nums_dict[value] == 1:
            return value
'''
# Key points: use dict, hashmap

# Intersection of Two Arrays II
'''
def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
    # use hashmp
    nums1_dict = {}
    res = []
    for num1 in nums1:
        if num1 in nums1_dict:
            nums1_dict[num1]+=1
        else:
            nums1_dict[num1]=1
    for num2 in nums2:
        if num2 in nums1_dict and nums1_dict[num2]!=0:
            res.append(num2)
            nums1_dict[num2]-=1
    return res
'''
# Key points: use dict, hashmap

#Plus One
'''
def plusOne(self, digits: List[int]) -> List[int]:
    i = len(digits)-1
    while i >= 0:
        if digits[i]+1<10:
            digits[i]=digits[i]+1
            return digits
        else:
            digits[i]=(digits[i]+1)%10
            i-=1
    # add one more digit 
    return [1]+digits
'''
# Key points: start from i, then i will move left 

# Move Zeroes
'''
def moveZeroes(self, nums: List[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    i = 0
    j = 0
    while j < len(nums):
        if nums[j]!=0:
            nums[i]=nums[j]
            i+=1
        j+=1
    nums[i:len(nums)]=[0]*(len(nums)-i)
'''
# Key points: get the non-zero elements first, then append the zero list 


# Two sum
'''
def twoSum(self, nums: List[int], target: int) -> List[int]:
    prevmap = {}
    for index, value in enumerate(nums):
        diff = target-value
        if value in prevmap:
            return [prevmap[value], index]
        prevmap[diff]=index
'''
# Use a hashmap to store the target-value

# Valid Sudoku
'''
def isValidSudoku(self, board: List[List[str]]) -> bool:
    res = set()
    for i in range(9):
        for j in range(9):
            ele = board[i][j]
            if ele != ".":
                row_cond = ele+" in row"+str(i)
                col_cond = ele+" in col"+str(j)
                sub_cond = ele+" in sub"+str(i//3)+str(j//3)
                if row_cond in res or col_cond in res or sub_cond in res:
                    return False
                else:
                    res.add(row_cond)
                    res.add(col_cond)
                    res.add(sub_cond)
    return True
'''
# Key points: record if a certain digit has appeared in a certain row, col or sub_block before

# Rotate Image
'''
def rotate(self, matrix: List[List[int]]) -> None:
    """
    Do not return anything, modify matrix in-place instead.
    """
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            if i < j:
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
        matrix[i].reverse()
'''
# Key points: flip first, then reverse row 


#Reverse String
'''
def reverseString(self, s: List[str]) -> None:
    """
    Do not return anything, modify s in-place instead.
    """
    i = 0
    j = len(s)-1
    while i < j:
        s[i], s[j] = s[j], s[i]
        i += 1
        j -= 1
'''
# Key points: move in two directions

# Reverse Integer
'''
def reverse(self, x: int) -> int:
    reverse_x = 0
    if x >= 0:
        sign = 1
    else:
        sign = -1
    x = abs(x)
    while x > 0:
        res = x%10
        x = x//10
        reverse_x = reverse_x*10+res
    reverse_x *= sign
    if reverse_x < -pow(2, 31) or reverse_x>pow(2, 31)-1:
        return 0
    else:
        return reverse_x
'''
# Key points: quotient and remainder

#First Unique Character in a String
'''
def firstUniqChar(self, s: str) -> int:
    count_dict = {}
    for d in s:
        if d not in count_dict:
            count_dict[d]=1
        else:
            count_dict[d]+=1
    for i in range(len(s)):
        if count_dict[s[i]]==1:
            return i
    return -1
'''
# Key points: store info in dict 

#Valid Anagram
'''
def isAnagram(self, s: str, t: str) -> bool:
    from collections import Counter
    return Counter(s)==Counter(t)
'''

#Valid Palindrome
'''
def isPalindrome(self, s: str) -> bool:
    i = 0
    j = len(s)-1
    while i < j:
        while not s[i].isalnum() and i<j:
            i += 1
        while not s[j].isalnum() and i<j:
            j-=1
        if s[i].lower() != s[j].lower():
            return False
        i+=1
        j-=1
    return True
'''
# Key points: isalnum, to add the main while into the separate while "i<j"

#String to Integer (atoi)
'''
def myAtoi(self, s: str) -> int:
    if len(s)==0:
        return 0

    sign = 1
    res = 0
    i = 0
    
    while s[i].isspace() and i < len(s)-1:
        i+=1
    if i > len(s)-1:
        return 0
    elif s[i].isalpha() or s[i]==".":
        return 0
    elif s[i] == "-":
        sign = -1
        i+=1
    elif s[i] == "+":
        sign = 1
        i+=1
    while i <= len(s)-1:
        if s[i].isdigit():
            res = res*10+int(s[i])
        else:
            break
        i+=1
    if sign*res < -pow(2,31):
        return -pow(2, 31)
    if sign*res > pow(2, 31)-1:
        return pow(2, 31)-1
    return sign*res
'''
# Key points:corner cases: len(s)==0, after a while loop, need to check i value again


# Implement strStr()
'''
def strStr(self, haystack: str, needle: str) -> int:
    # corner cases
    if len(needle)==0:
        return 0
    if len(haystack)==0 and len(needle)!=0:
        return -1

    i = 0
    while i <= len(haystack)-1:
        # try to find the first matching element 
        if haystack[i]!=needle[0]:
            i+=1
        else:
            # check if we can get the needle
            if i+len(needle)>len(haystack):
                return -1
            else:
                if haystack[i:i+len(needle)] == needle:
                    return i
                # if not, move i one more step
                else:
                    i+=1
    return -1
'''

# Count and Say
'''
def countAndSay(self, n: int) -> str:
    if n == 1:
        return "1"
    seq = "1"
    for i in range(n-1):
        seq = self.getnext(seq)
    return seq

def getnext(self, seq):
    i = j = 0
    seq = list(seq)
    res = []
    seq.append(None)
    while i < len(seq)-1:
        j += 1
        if seq[i] != seq[j]:
            res += [j-i, seq[i]]
            i = j
    return "".join(str(s) for s in res)
'''

# Longest Common Prefix
'''
def longestCommonPrefix(self, strs: List[str]) -> str:
    if len(strs)==1:
        return strs[0]
    min_len = min([len(st) for st in strs])
    if min_len == 0:
        return ""
    i = 1
    while i <= min_len:
        if len(set(st[:i] for st in strs))==1:
            i+=1
        else:
            break
    return strs[0][:i-1]
'''
# Key points: corner_cases, break from while loop and then return 

#Delete Node in a Linked List
'''
def deleteNode(self, node):
    """
    :type node: ListNode
    :rtype: void Do not return anything, modify node in-place instead.
    """
    # can access the node directly
    curnode = node
    # before getting the tail of the linkedlist
    while curnode.next is not None:
        temp_node = curnode
        # Put cornode.next in curnode
        curnode.val = curnode.next.val
        curnode = curnode.next
    # remove the last node
    temp_node.next = None
'''

#Remove Nth Node From End of List
'''
def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
    slow = fast = head
    # fast will move n steps ahead
    for i in range(n):
        fast = fast.next
    # when it has exceeds the length
    if not fast:
        # remove head
        return head.next
    #before reaching the tail
    while fast.next is not None:
        slow = slow.next
        fast = fast.next
    slow.next = slow.next.next
    return head
'''
# Key points: slow and fast, fast is n step ahead


# Reverse Linked List
'''
def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    curnode = head
    prevnode = None
    while curnode is not None:
        # it will change, so save it temporarily
        temp_node = curnode.next
        curnode.next = prevnode
        # move along the linkedlist
        prevnode = curnode
        curnode = temp_node
    return prevnode
'''

# Merge Two Sorted Lists
'''
def mergeTwoLists(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    # create a head
    res_node = ListNode(0)
    fixed_head = res_node
    while l1 is not None and l2 is not None:
        if l1.val > l2.val:
            res_node.next = l2
            l2 = l2.next
        else:
            res_node.next = l1
            l1 =  l1.next
        res_node = res_node.next

    if l1:
        res_node.next = l1
    if l2:
        res_node.next = l2
    return fixed_head.next
'''


#Palindrome Linked List
'''
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        # directly use list
        vals = []
        while head is not None:
            vals+=[head.val]
            head = head.next
        return vals == vals[::-1]
'''

#Linked List Cycle
'''
def hasCycle(self, head: Optional[ListNode]) -> bool:
    slow = fast = head
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
        # If there is a circle
        if slow==fast:
            return True
    return False
'''

# Maximum Depth of Binary Tree
'''
def maxDepth(self, root: Optional[TreeNode]) -> int:
    res = []
    self.level(root, 0, res)
    return len(res)

def level(self, node, level, res):
    if not node:
        return

    if level >= len(res):
        # Have not reached this level before
        res.append([])
    # add node.val to current level
    res[level].append(node.val)
    # add node.left and node.right 
    self.level(node.left, level+1, res)
    self.level(node.right, level+1, res)
'''


#Merge Sorted Array
'''
def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
    """
    Do not return anything, modify nums1 in-place instead.
    """
    # three pointers
    p1, p2, p = m-1, n-1, m+n-1
    while p1 >= 0 and p2>=0:
        if nums1[p1]>=nums2[p2]:
            nums1[p]=nums1[p1]
            p1-=1
            p-=1
        else:
            nums1[p]=nums2[p2]
            p2-=1
            p-=1
    # to deal with the cases when remaining num in nums2 are lower
    nums1[:p2+1] = nums2[:p2+1]
'''

#Validate Binary Search Tree
'''
def isValidBST(self, root: Optional[TreeNode]) -> bool:
    # for every node, we can find min and max
    def isValid(root,lessthan,morethan):
        if not root:
            return True
        if root.val >= lessthan or root.val <= morethan:
            return False

        return (isValid(root.left,root.val,morethan) and isValid(root.right,lessthan,root.val))

    return isValid(root,float("inf"),float("-inf"))
'''

#Symmetric Tree
'''
def isSymmetric(self, root: Optional[TreeNode]) -> bool:
    # recursive
    if not root:
        return True
    return self.isSym_helper(root.left, root.right)

def isSym_helper(self, left_node, right_node):
    if not left_node and not right_node:
        return True

    if not right_node or not left_node or left_node.val != right_node.val:
        return False

    return self.isSym_helper(left_node.left, right_node.right) and self.isSym_helper(left_node.right, right_node.left)
'''

# Binary Tree Level Order Traversal
'''
def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
    res = []
    self.level(root, 0, res)
    return res

def level(self, node, level, res):
    if not node:
        return

    if level >= len(res):
        # Have not reached this level before
        res.append([])
    res[level].append(node.val)
    self.level(node.left, level+1, res)
    self.level(node.right, level+1, res)
'''
# Same as Maximum Depth of Binary Tree

#Convert Sorted Array to Binary Search Tree
'''
def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
    def helper(nums, l, r):
        if l > r:
            return None
        # first build the node
        mid=(l+r)//2
        node = TreeNode(nums[mid])
        # left side
        node.left = helper(nums, l, mid-1)
        # right side
        node.right = helper(nums, mid+1, r)
        return node

    return helper(nums, 0, len(nums)-1)
'''

# First Bad Version
'''
def firstBadVersion(self, n):
    """
    :type n: int
    :rtype: int
    """
    l = 1
    r = n
    while l < r:
        mid = (l+r)//2
        if isBadVersion(mid):
            r = mid
        else:
            l = mid+1
    return l       
'''


# Climbing Stairs
'''
def climbStairs(self, n: int) -> int:
    dp=n*[0]
    if n==1:
        return 1
    if n==2:
        return 2
    dp[0]=1
    dp[1]=2
    i = 2
    while i <= n-1:
        dp[i]=dp[i-1]+dp[i-2]
        i+=1
    return dp[n-1]
'''


#Best Time to Buy and Sell Stock
'''
def maxProfit(self, prices: List[int]) -> int:
    # one pass
    min_price = prices[0]
    max_profit = 0

    for i in range(len(prices)):
        if prices[i] < min_price:
            min_price = prices[i]
        else:
            if prices[i]-min_price>max_profit:
                max_profit = prices[i]-min_price
    return max_profit
'''

#Maximum Subarray
'''
def maxSubArray(self, nums: List[int]) -> int:
    maxarray = curarray = nums[0]

    for num in nums[1:]:
        curarray = max(num, curarray+num)
        maxarray = max(curarray, maxarray)
    return maxarray
'''

#House Robber
'''
def rob(self, nums: List[int]) -> int:
    if not nums:
        return 0

    dp = [0, nums[0]]

    for i in range(2, len(nums)+1):
        dp.append(max(dp[i-1], dp[i-2]+nums[i-1]))
    return dp[len(nums)]      
'''

#Fizz Buzz
'''
def fizzBuzz(self, n: int) -> List[str]:
    res = []
    for i in range(1, n+1):
        if i%3 == 0 and i%5 == 0:
            res.append("FizzBuzz")
        elif i%3 == 0:
            res.append("Fizz")
        elif i%5 == 0:
            res.append("Buzz")
        else:
            res.append(str(i))
    return res
'''

#Count Primes
'''
def countPrimes(self, n: int) -> int:
    if n <= 2:
        return 0
    res = [True] * n
    res[0] = res[1] = False
    for i in range(2, n):
        if res[i] == True:
            for j in range(i, (n-1)//i+1):
                res[i*j] = False
    return sum(res)
'''

#Power of Three
'''
def isPowerOfThree(self, n: int) -> bool:
    i = 1
    while i <= n:
        if i == n:
            return True
        i = i*3
    return False
'''

#Roman to Integer
'''
def romanToInt(self, s: str) -> int:
    dic = {"I": 1,"V": 5,"X":10,
     "L":50,"C":100,"D":500,"M":1000, None:10000}
    pre = None
    ans=0
    for i in s:
        if dic[i]<=dic[pre]:
            ans+=dic[i]
        else:
            ans+=(dic[i]-2*dic[pre])
        pre=i
    return ans
'''
# set pre as None first

#  Number of 1 Bits
'''
def hammingWeight(self, n: int) -> int:
    count = 0

    if not n:
        return 0

    while n:
        count += n&1
        n  = n >>1
    return count
'''

# Hamming Distance
'''
def hammingDistance(self, x: int, y: int) -> int:
    t = x ^ y
    count = 0
    while t:
        count += t & 1
        t = t >> 1
    return count
'''

# Reverse Bits
'''
def reverseBits(self, n: int) -> int:
    s = 0
    for i in range(32):
        num = n % 2
        s = s*2+num
        n = n>>1

    return s
'''

#Pascal's Triangle
'''
    def generate(self, numRows: int) -> List[List[int]]:
        final_res = [[1], [1,1]]
        if numRows == 1:
            return [[1]]
        if numRows == 2:
            return [[1], [1,1]]
        cur_row = [1,1]
        for i in range(3, numRows+1):
            cur_row = self.getnextrow(cur_row)
            final_res.append(cur_row)
        return final_res
                  
    def getnextrow(self, cur_row):
        res = []
        for i in range(len(cur_row)-1):
            res.append(cur_row[i]+cur_row[i+1])
        res = [1]+res+[1]
        return res
'''

#Valid Parentheses
'''
def isValid(self, s: str) -> bool:
    match_dict = {
        ")":"(",
        "]":"[",
        "}":"{"
    }
    stack = []
    for d in s:
        if d in match_dict:
        # Stack: last in, first out 
            top_element = stack.pop() if stack else "#"
            if top_element != match_dict[d]:
                return False
        else:
            stack.append(d)
    return not stack
'''


# Missing Number
'''
def missingNumber(self, nums: List[int]) -> int:
    exp_sum = (1+len(nums))*len(nums)//2
    real_sum = sum(nums)
    return exp_sum-real_sum
'''

## Top Interview Questions - Medium
# 3Sum
'''
    def threeSum(self, nums: List[int]) -> List[List[int]]:

        n = len(nums)
        nums=sorted(nums)
        res = []
        
        if n<=2:
            return res
        
        for i in range(n-2):
            if i>0 and nums[i]==nums[i-1]:
                continue #jump out of iteration
            new_sum=-nums[i]
            
            j=i+1
            k=n-1
            while j<k:
                if nums[j]+nums[k]>new_sum:
                    k-=1
                elif nums[j]+nums[k]<new_sum:
                    j+=1
                else:
                    res.append([nums[i], nums[j], nums[k]])
                    while j<k and nums[j]==nums[j+1]:
                        j+=1
                    j+=1
                    while j<k and nums[k]==nums[k-1]:
                        k-=1
                    k-=1
        return res
'''
# faster

'''
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums = sorted(nums)
        if len(nums)<=2:
            return []
        res = []
        for i in range(len(nums)-2):
            target = -nums[i]
            hashmap = {}
            for j in range(i+1, len(nums)):
                if nums[j] in hashmap:
                    res_ls = [nums[i], nums[hashmap[nums[j]]], nums[j]]
                    if res_ls not in res:
                        res.append(res_ls)
                hashmap[target-nums[j]] = j

        return res
'''

# Set Matrix Zeroes
'''
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        rows = []
        cols = []
        i = j = 0
        while i < len(matrix):
            while j < len(matrix[0]) and i < len(matrix):
                if matrix[i][j]==0:
                    rows.append(i)
                    cols.append(j)
                    j+=1
                else:
                    j+=1
            j=0
            i+=1
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if i in rows or j in cols:
                    matrix[i][j] = 0
'''

# Group Anagrams
'''
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        res = []
        hashmap = {}
        for st in strs:
            if str(sorted(st)) not in hashmap:
                hashmap[str(sorted(st))] = [st]
            else:
                hashmap[str(sorted(st))].append(st)
        for key in hashmap:
            res.append(hashmap[key])
        return res
'''

# Longest Substring Without Repeating Characters
'''
def lengthOfLongestSubstring(self, s: str) -> int:
    if len(s)<=1:
        return len(s)
    max_len = 0
    i = 0
    j = 1
    while j < len(s):
        if s[j] not in s[i:j]:
            max_len = max(j-i+1, max_len)
            j+=1
        else:
            i+=1      
    return max_len
'''

# Longest Palindromic Substring
'''
    def longestPalindrome(self, s: str) -> str:
        res = ""
        for i in range(len(s)):
            tmp1 = self.helper(s, i, i)
            tmp2 = self.helper(s,i,i+1)
            if len(tmp1)>len(res):
                res=tmp1
            if len(tmp2)>len(res):
                res=tmp2
        return res
        
        
        
    def helper(self, s, l, r):
        while l>=0 and r<=len(s)-1 and s[l]==s[r]:
            l-=1
            r+=1
        return s[l+1:r]
'''


# Increasing Triplet Subsequence
'''
    def increasingTriplet(self, nums: List[int]) -> bool:
        small_1st = max(nums)+1
        small_2nd = max(nums)+1
        for n in nums:
            if n<=small_1st:
                small_1st=n
            elif n<=small_2nd:
                small_2nd=n
            else:
                return True
        return False
'''

#Missing Ranges
'''
def findMissingRanges(self, nums: List[int], lower: int, upper: int) -> List[str]:
    if len(nums)==0:
        if lower < upper:
            return [str(lower)+"->"+str(upper)]
        else:
            return [str(lower)]


    res = []
    # add 1st and last element
    if nums[0] > lower:
        nums = [lower-1]+nums
    if nums[-1] < upper:
        nums.append(upper+1)

    for i in range(len(nums)-1):
        if nums[i]+1 != nums[i+1]:
            min_value = min(nums[i]+1, upper)
            max_value = max(nums[i+1]-1, lower)
            if min_value == max_value:
                res.append(str(min_value))
            else:
                res.append(str(min_value)+"->"+str(max_value))
    return res
'''

#Add Two Numbers
'''
def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    # special cases
    if l1 == None:
        return l2
    if l2 == None:
        return l2
    # create head, put dummy static
    head = ListNode(0)
    dummy = head
    add = 0
    while l1 or l2:
        l1_val = l1.val if l1 else 0
        l2_val = l2.val if l2 else 0
        res = (add+l1_val+l2_val)%10
        add = (add+l1_val+l2_val)//10
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
        # create next node, and then move to next node
        head.next = ListNode(res)
        head = head.next
    if add:
        head.next = ListNode(add)
    return dummy.next
'''

#Odd Even Linked List
'''
def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    dummy1=odd=ListNode(0)
    dummy2=even=ListNode(0)

    while head:
        odd.next=head
        even.next=head.next
        odd = odd.next
        even = even.next
        head=head.next.next if even else None
    odd.next =dummy2.next
    return dummy1.next
'''

#Intersection of Two Linked Lists
'''
def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
    headB_set = set()

    while headB:
        headB_set.add(headB)
        headB = headB.next

    while headA:
        if headA in headB_set:
            return headA
        headA = headA.next
    return None
'''

#Letter Combinations of a Phone Number
'''
def letterCombinations(self, digits: str) -> List[str]:
    if len(digits)==0:
        return []
    letter_map={"2":["a","b","c"],
                "3":["d","e","f"],
                "4":["g","h","i"],
                "5":["j","k","l"],
                "6":["m","n","o"],
                "7":["p","q","r","s"],
                "8":["t","u","v"],
                "9":["w","x","y","z"]}
    def backtrack(index, path):
        if len(path)==len(digits):
            combinations.append("".join(path))
            return 
        possible_letters = letter_map[digits[index]]
        for letter in possible_letters:
            path.append(letter)
            backtrack(index+1,path)
            # removing the letter
            path.pop()

    combinations = []
    backtrack(0, [])
    return combinations
'''

#Generate Parentheses
'''
def generateParenthesis(self, n: int) -> List[str]:
    ans=[] 
    def backtrack(s=[], left=0, right=0):
        if len(s)==2*n:
            ans.append("".join(s))
            return 
        if left < n:
            s.append("(")
            backtrack(s, left+1, right)
            s.pop()
        if right < left:
            s.append(")")
            backtrack(s,left,right+1)
            s.pop()


    backtrack()
    return ans
'''

#Permutations
'''
def permute(self, nums: List[int]) -> List[List[int]]:

    # use ith element, in the index location

    def backtrack(first=0):
        if first == n:
            output.append(nums[:])
            return
        for i in range(first, n):
            nums[first],nums[i] = nums[i], nums[first]
            # deal with the next element 
            backtrack(first+1)
            # backtrack
            nums[first],nums[i] = nums[i], nums[first]
    n = len(nums)
    output =[]
    backtrack()
    return output
'''

#Subsets
'''
def subsets(self, nums: List[int]) -> List[List[int]]:
    n = len(nums)
    output = [[]]
    for num in nums:
        output+=[curr + [num] for curr in output]
    return output 
'''

#Sort Colors
'''
def sortColors(self, nums: List[int]) -> None:
    """
    Do not return anything, modify nums in-place instead.
    """
    hashmap = {}
    res = []
    for num in nums:
        if num not in hashmap:
            hashmap[num] = 1
        else:
            hashmap[num]+=1
    for key in [0, 1, 2]:
        if key in hashmap:
            res+=[key]*hashmap[key]
    nums[:] = res
'''

#Top K Frequent Elements
'''
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        from collections import Counter
        nums_count = Counter(nums)
        key_value_pair = []
        res = []
        for key, value in nums_count.items():
            key_value_pair.append([value, key])
        key_value_pair = sorted(key_value_pair, reverse=True)
        for i in range(0,k):
            res.append(key_value_pair[i][1])
        return res
'''

#Kth Largest Element in an Array
'''
def findKthLargest(self, nums: List[int], k: int) -> int:
    sorted_nums = sorted(nums, reverse=True)
    return sorted_nums[k-1]
'''

#Find Peak Element
'''
def findPeakElement(self, nums: List[int]) -> int:
    left = 0
    right = len(nums)-1

    while left < right:
        mid = (left+right)//2
        if nums[mid] > nums[mid+1]:
            right = mid
        else:
            left = mid+1
    return left 
'''

#Search for a Range
'''
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if len(nums)==0:
            return [-1,-1]
        
        res_0 = self.findfirst(nums, target)
        res_1 = self.findlast(nums, target)
        return [res_0, res_1]
        
    def findfirst(self, nums, target):
            left = 0
            right = len(nums)-1
            while left<right:
                mid = (left+right)//2
                if nums[mid]<target:
                    left=mid+1
                elif nums[mid]>target:
                    right=mid-1
                else:
                    right=mid
            if nums[left]==target:
                return left
            else:
                return -1
         
    def findlast(self,nums, target):
            left = 0
            right = len(nums)-1
            while left<right:
                mid = (left+right+1)//2
                if nums[mid]<target:
                    left=mid+1
                elif nums[mid]>target:
                    right=mid-1
                else:
                    left=mid
            if nums[right]==target:
                return right
            else:
                return -1
'''

#Merge Intervals
'''
def merge(self, intervals: List[List[int]]) -> List[List[int]]:
    res =[]
    for interval in sorted(intervals, key=lambda x: x[0]):
        if not res or res[-1][1]<interval[0]:
            res.append(interval)
        else:
            res[-1][1] = max(res[-1][1],interval[1])
    return res
'''

#Search in Rotated Sorted Array
'''
    def search(self, nums: List[int], target: int) -> int:
        if len(nums) == 0:
            return -1
        start = 0
        end = len(nums)-1
        while start <= end:
            mid = (start+end)//2
            if nums[mid] == target:
                return mid
            else:
                if nums[mid]  <= nums[end]:
                    if target >= nums[mid] and target <= nums[end]:
                        start = mid +1
                    else:
                        end = mid -1
                else:
                    if target >= nums[start] and target <= nums[mid]:
                        end = mid -1
                    else:
                        start = mid+1
        return -1            
'''

#Jump Game
'''
def canJump(self, nums: List[int]) -> bool:
    if len(nums)==0:
        return False
    # go back
    i = len(nums)-1
    target = len(nums)-1
    while i >= 0:
        if i + nums[i] >= target:
            # can achieve target
            target = i
        i-=1
    if target == 0:
        return True
    else:
        return False
'''

#Unique Paths
'''
    def uniquePaths(self, m: int, n: int) -> int:
        if m < 0 or n <0:
            return 0
        row = [0]*(m+1)
        f = [row for i in range(n+1)]
        f[1][1]=1
        for i in range(1, n+1):
            for j in range(1, m+1):
                f[i][j] = f[i-1][j]+f[i][j-1]
        return f[n][m]
'''

#Coin Change
'''
def coinChange(self, coins: List[int], amount: int) -> int:
    dp = (amount+1)*[float("inf")]
    dp[0]=0
    for i in range(amount+1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i-coin]+1)

    if dp[amount] > amount:
        return -1
    else:
        return dp[amount]
'''

```
